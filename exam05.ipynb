```python
from pandas import DataFrame
import numpy as np
```

```python
df1 = DataFrame(np.zeros((3,3),dtype='int32'),index=['one','two','three'])
df1
```

```python
df2 = df1 + 3
df2
```

```python
df3 = DataFrame(np.ones((3,3),dtype='int32'),index=['one','two','three'])
df3
```

```python
df4 = df2 + df3
df4
```

```python
df5 = DataFrame(np.ones((4,4),dtype='int32'),index=['one','two','three','four'])
df5
```

```python
df6 = df4 + df5
df6
```

```python
df4.add(df6, fill_value=0)
```
